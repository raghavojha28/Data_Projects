```python
#importing the modules that i'll be using in this project

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure 

%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (20,10)

pd.options.mode.chained_assignment = None
```

```python
# Reading through the CSV file 
df = pd.read_csv(r"D:\Movie Industy Dataset\movies.csv")
```

```python
df.head()
```

```python
# Checking missing data
# to do this, Loop through the data and see if their is anything missing

for col in df.columns:
    missing_percent = np.mean(df[col].isnull())
    print ('{}: {}%'.format(col, missing_percent))
```

```python
df['budget'] = df['budget'].fillna(value=0)
df['gross'] = df['gross'].fillna(value=0)
df['company'] = df['company'].fillna(value=0)
df['votes'] = df['votes'].fillna(value=0)
```

```python
# Checking Data Types for our columns

print(df.dtypes)
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df.head()
```

```python
# Now we want to make sure the year in the "year" column and "released" column should match.
# Adding new column called yearcorrect

df['yearcorrect'] = df['released'].astype(str).str.extract(r'\b(\d{4})\b')
df['yearcorrect'].fillna(value=0).astype('int64')
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace= False, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Correlating Budget Vs Gross

plt.scatter(x = df['budget'], y=df['gross'])

plt.title('Budget vs Gross Revenue')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Revenue')

plt.show()
```

```python
#Looking for Correlation Between Company vs Gross

df.head()
```

```python
# Since company is 'object' dtype. 
# We are going to convert all 'object' datatype into 'category' to make a correlation plot.


df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
```

```python
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True, method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Between Numeric Metrics')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
# Looking at the Highest Correlations with each matrix

corr_mat = df_numerized.corr()

corr_pairs = corr_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
