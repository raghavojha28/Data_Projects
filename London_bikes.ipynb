```python
# putting the column names that i want to use

col_dict = {
    "timestamp" : "time",
    "cnt" : "count",
    "t1" : "temp_real_c", 
    "t2" : "temp_feels_like_c",
    "hum" : "humidity_percent",
    "wind_speed" : "wind_speed_kph", 
    "weather_code" : "weather", 
    "is_holiday" : "is_holiday", 
    "is_weekend" : "is_weekend", 
    "season" : "season",
}

# renaming the column names with the specified dict
df.rename(col_dict, axis=1, inplace= True)
```

```python
df.head()
```

```python
# changing the humidity percent to values(i.e. a value between 0 and 1)
df.humidity_percent = df.humidity_percent / 100
```

```python
# creating the season dictionary to map the numbers from 0-3 to the actual seasons

season_dict = {
    "0.0" : "spring", 
    "1.0" : "summer", 
    "2.0" : "autumn", 
    "3.0" : "winter" 
}

# creating the weather dictionary so that we can map the values to actual weather
weather_dict = {
    "1.0" : "clear", 
    "2.0" : "scattered clouds", 
    "3.0" : "broken clouds", 
    "7.0" : "cloudy", 
    "4.0" : "Rain", 
    "10.0" : "Rain with thunderstorm", 
    "26.0" : "snowfall" 
}
```

```python
# changing the data type of season column to string
df.season = df.season.astype("str")

# mapping the values to the atucal season
df.season = df.season.map(season_dict)

# # changing the data type of weather column to string
df.weather = df.weather.astype('str')

# mapping the weather codes to the actual weather names
df.weather = df.weather.map(weather_dict)
```
